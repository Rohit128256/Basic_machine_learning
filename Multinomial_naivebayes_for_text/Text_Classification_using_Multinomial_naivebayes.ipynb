```python
dict = {}
for i in range(len(x_train)):
    #coverting the each article(type str) to lower case
    arti = x_train[i][1].lower() #that arti is string
    #now split the article into list of words 
    strp_words = re.split(r'\W+',arti)
    # print(strp_words)
    #iterating over each word and adding it tio dictionary with it's count
    for s in strp_words:
        #we will not include stop_words, alpha-numerics, punctuations or irrelevant word of length less than 2 in our dictionary
        if not(s.isalpha()) or (s in stop_word) or (len(s)<=2):
            continue
        if s in dict:
            dict[s]+=1
        else:
            dict[s]=1
dict
```

```python
effective_words = list(dct2.keys())[:6500]

# making the function to make training datasets
def makedataset(x_tr_te,features):
    x_train_dataset = np.zeros([len(x_tr_te),len(features)],int)
    for i in range(len(x_tr_te)):
        art_str = x_tr_te[i][1].lower()
        lst_words = re.split(r'\W+',art_str)
        for j in lst_words:
            if j in features:
                x_train_dataset[i][features.index(j)] += 1
    return x_train_dataset

#making the train data set
x_train_dataset = makedataset(x_train,effective_words)
x_test_dataset = makedataset(x_test,effective_words)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
clf = MultinomialNB()
clf.fit(x_train_dataset,y_train)
y_pred = clf.predict(x_test_dataset)
```

```python
print("The score on train data -",clf.score(x_train_dataset,y_train))
print("The score on test data -",clf.score(x_test_dataset,y_test))
# print("The confusion matrix for test data is -> ",confusion_matrix(y_test,y_pred))
print("The classification report ->",classification_report(y_test,y_pred))
```

```python
# Implementing the code for myself 
def fit(x_train_dataset,features,y_train):
    dct = {}
    y_train=np.array(y_train)
    cls_lst = set(y_train)
    dct['Total_docs'] = len(y_train)
    for cls in cls_lst:
        temp = 0
        dct[cls]={}
        x_train_curr_cls = x_train_dataset[y_train == cls]
        for fea in features:
            dct[cls][fea] = (x_train_curr_cls[:,features.index(fea)]).sum()
            temp+=dct[cls][fea]
        dct[cls]['total_words_in_clss'] = temp
        dct[cls]['Length'] = x_train_curr_cls.shape[0]
    return dct

dctt = fit(x_train_dataset,effective_words,y_train)
print(dctt['alt.atheism']['total_words_in_clss'])
print(sum(list(dctt['alt.atheism'].values()))-((dctt['alt.atheism']['Length'])+(dctt['alt.atheism']['total_words_in_clss'])))
```

```python

#def the probablity function:
def probability(cls,datapoint,dict):
    cls_p = np.log(dict[cls]['Length'])-np.log(dict['Total_docs'])
    lst_fea = list(dict[cls].keys())
    for i in lst_fea:
        if (i == 'Length') or (i == 'total_words_in_clss'):
            continue
        elif(datapoint[lst_fea.index(i)] == 0):
            prob_curr_fea = 0
        else:
            count_fea_clss = dict[cls][i] + 1
            count_total_words = dict[cls]['total_words_in_clss']+len(dict[cls].keys())-2
            prob_curr_fea = np.log(count_fea_clss) - np.log(count_total_words)
        cls_p += prob_curr_fea
    return cls_p

```
