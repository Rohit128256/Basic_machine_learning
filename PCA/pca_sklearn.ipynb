```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
np.random.seed(6456546)
```

```python
mean_vec_1 = np.array([0,0,0])
cov_mat1 = np.array([[1,0,0],[0,1,0],[0,0,1]])
class_1 = np.random.multivariate_normal(mean_vec_1,cov_mat1,100)
```

```python
mean_vec_2 = np.array([1,1,1])
cov_mat2 = np.array([[1,0,0],[0,1,0],[0,0,1]])
class_2 = np.random.multivariate_normal(mean_vec_2,cov_mat2,100)
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111,projection='3d')
ax.plot(class_1[:,0],class_1[:,1],class_1[:,2],'o')
ax.plot(class_2[:,0],class_2[:,1],class_2[:,2],'^')
plt.show()
```

```python
all_data = np.concatenate((class_1,class_2))
```

```python
pca = PCA()
t_data_all = pca.fit_transform(all_data)
transformed_data = t_data_all[:,0:2]
transformed_data.shape
```

```python
pca.components_ , pca.explained_variance_
```

```python
plt.plot(transformed_data[:100,0],transformed_data[:100,1],'o')
plt.plot(transformed_data[100:,0],transformed_data[100:,1],'^')
plt.show()
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111,projection='3d')
ax.plot(t_data_all[:,0],t_data_all[:,1],t_data_all[:,2],'^')
plt.show()
```
