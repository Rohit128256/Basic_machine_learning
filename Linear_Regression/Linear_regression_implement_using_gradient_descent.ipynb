```python
import numpy as np
from sklearn import datasets
data = datasets.load_digits()

#define cost or error per data function
def cost(x,y,m,c):
    n = len(x)
    mean_cost = 0
    for i in range(n):
        mean_cost += (1/n)*((y[i]-(m*x[i]).sum()-c)**2)
    return mean_cost

#define the step gradient function 
def step(x,y,a,m,c):
    cols = x.shape[1]
    rows = x.shape[0]
    # m_slope and c_slope will be an array here
    m_slope = []
    c_slope = 0
    for j in range(cols):
        m_slope_for_each = 0
        for i in range(rows):
            m_slope_for_each += ((-2/rows)*(y[i]-((m*x[i]).sum())-c)*x[i,j])
            c_slope += ((-2/rows)*(y[i]-((m*x[i]).sum())-c))
        m_slope.append(m_slope_for_each)
    m_slope = np.array(m_slope)
    m_new = m - a*m_slope
    c_new = c - a*c_slope
    return m_new, c_new

#define the gradient function
def grd(x,y,a,i):
    m = np.zeros(x.shape[1],dtype=float)
    c = 0.0
    for i in range(i):
        m , c = step(x,y,a,m,c)
        print("cost in ",i+1,"iteration",cost(x,y,m,c))
    return m, c

l_rate = 0.0003
itter = 150
x = data.data
y = data.target
m, c = grd(x,y,l_rate,itter)
print("The Final m & c is ->",m,"and",c)
```
