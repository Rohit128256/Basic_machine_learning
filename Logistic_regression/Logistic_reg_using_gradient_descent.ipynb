```python
import numpy as np
from sklearn import datasets
from sklearn import model_selection
br_c = datasets.load_breast_cancer()
x = br_c.data
y = br_c.target
x_train,x_test,y_train,y_test = model_selection.train_test_split(x,y,train_size=0.8)
print(x_train.shape,x_test.shape)
# def prediction function
def pred(x,m,c):
    y_pred = []
    y_proba = []
    rows = x.shape[0]
    for i in range(rows):
        p = h(x[i],m,c)
        y_proba.append(p)
        if(p>0.5):
            y_pred.append(1)
        else:
            y_pred.append(0)
    y_pred = np.array(y_pred)
    y_proba = np.array(y_proba)
    return y_pred, y_proba


# def cost per point function:
def cost(x,y,m,c):
    rows = x.shape[0]
    m_cost = 0
    for i in range(rows):
        m_cost += np.log(1 + ((np.e)**(np.dot(x[i],m)+c))) - y[i]*np.dot(m,x[i])
    return m_cost/rows

# def sigmoid function
def h(x,m,c):
    s = 1/(1+((np.e)**(-(np.dot(x,m)+c))))
    return s

# def step gradient function
def stepping(x,y,a,m,c):
    rows = x.shape[0]
    cols = x.shape[1]
    m_slope = []
    c_slope = 0
    for j in range(cols):
        m_temp = 0
        for i in range(rows):
            m_temp += (-1/rows)*(y[i] - h(x[i],m,c))*x[i,j]
            c_slope += (-1/rows)*(y[i] - h(x[i],m,c))
        m_slope.append(m_temp)
    m_slope = np.array(m_slope)
    m_new = m - a*m_slope
    c_new = c - a*c_slope
    return m_new , c_new


# def gradient descent func
def gd(x,y,a,i):
    m = np.zeros(x.shape[1],dtype=float)
    c = 0 
    for i in range(i):
        m,c = stepping(x,y,a,m,c)
        # print("The cost in ",i,"iterration is -> ",cost(x,y,m,c))
    return m,c

l_rate = 0.000005
itter = 800
m, c = gd(x_train,y_train,l_rate,itter)
print("the final values of m and c is -",m,"and",c)
y_pred, y_proba = pred(x_test,m,c)
print(y_pred-y_test)
y_proba
```
